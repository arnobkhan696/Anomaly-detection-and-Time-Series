1. What is Anomaly Detection? Explain its types (point, contextual, and
collective anomalies) with examples.
- Anomaly Detection is a process of identifying patterns in data that do not conform to expected behavior. These irregular patterns are called anomalies or outliers. Anomalies often indicate critical incidents, such as fraud, system faults, or errors in data.
- Types of Anomalies

  1️⃣ Point Anomalies (Global Anomalies)

  Definition: A single data point is considered anomalous if it deviates significantly from the rest of the data.

  Example:

  In a credit card transaction dataset, a purchase of $10,000 in a single transaction when the user usually spends $10–$100 is a point anomaly.

  In sensor data: A temperature reading of 150°C in a system that normally operates at 30–40°C.

  2️⃣ Contextual Anomalies (Conditional Anomalies)

  Definition: A data point is anomalous in a specific context, but not necessarily anomalous in general.

  Example:

  A temperature of 25°C is normal in summer but anomalous in winter for a given region.

  In time-series data: High electricity usage at midnight may be anomalous, while the same usage at 3 PM is normal.

  3️⃣ Collective Anomalies (Group Anomalies)

  Definition: A collection of related data points is anomalous together, even if individual points seem normal.

  Example:

  A sudden burst of multiple login attempts from different IPs in a short period indicates a collective anomaly (possible cyberattack).

  In a manufacturing process, a sequence of sensor readings showing a gradual drift indicates a malfunction.

2. Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.
 -  1️⃣ Isolation Forest (iForest)

 Approach:

Based on the idea that anomalies are easier to isolate than normal points.

Randomly partitions data by selecting features and split values.

Anomalies tend to have shorter paths in the tree structure (isolated in fewer splits).

 Suitable Use Cases:

High-dimensional datasets (e.g., fraud detection in credit card transactions).

When you need a fast, scalable method for anomaly detection.

Works well for detecting point anomalies in large datasets.

📋 Pros:

Efficient with large datasets.

No need for labeled data (unsupervised).

📋 Cons:

May not perform well on contextual or collective anomalies.

✅ 2️⃣ DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

🔍 Approach:

Clustering algorithm that groups data points based on density (number of nearby points).

Points in low-density regions are marked as outliers (noise).

Parameters:

eps: Maximum distance to consider neighbors.

min_samples: Minimum number of points required to form a dense region.

📊 Suitable Use Cases:

Spatial or geolocation data (e.g., detecting GPS outliers).

Situations where clusters have arbitrary shapes.

Detects both cluster structure and anomalies simultaneously.

📋 Pros:

No need to predefine the number of clusters.

Finds arbitrary-shaped clusters.

📋 Cons:

Sensitive to parameter selection.

May struggle with varying density data.

✅ 3️⃣ Local Outlier Factor (LOF)

🔍 Approach:

Measures local density deviation of a point compared to its neighbors.

A point is anomalous if its local density is much lower than its neighbors.

Provides an anomaly score (higher score = more anomalous).

📊 Suitable Use Cases:

Detecting local anomalies where global anomaly detection fails.

Time-series data or data where context matters (e.g., abnormal usage of a network at a specific time).

Useful for detecting contextual and point anomalies.

📋 Pros:

Captures local variations well.

Works well in datasets where data density varies.

📋 Cons:

Computationally expensive for large datasets.

Choice of neighborhood size (k) affects results.

   Conclusion:

Use Isolation Forest when you want fast anomaly detection in high-dimensional datasets focusing on global point anomalies.

Choose DBSCAN when clustering and anomaly detection in spatial data is important, especially with unknown cluster shapes.

Prefer LOF when local anomalies matter, especially in data with varying densities or contextual dependencies.

3. What are the key components of a Time Series? Explain each with one
example.
- A Time Series is a sequence of data points collected or recorded at regular time intervals. Time series analysis helps us understand the underlying patterns and make forecasts.

  Here are the four key components of a Time Series:

  1️⃣ Trend (T)

🔍 Definition:
A long-term increase or decrease in the data over time. It shows the general direction in which the time series is moving.

📊 Example:

The annual sales revenue of a company steadily increasing over several years indicates a positive trend.

E.g., Sales increased from $1M in 2015 to $3M in 2020.

2️⃣ Seasonality (S)

🔍 Definition:
A repeating pattern or cycle in data that occurs at fixed intervals, typically within a year, month, or week, caused by seasonal factors.

📊 Example:

Ice cream sales increase during summer and decrease in winter every year.

E.g., High sales every June–August.

3️⃣ Cyclic Component (C)

🔍 Definition:
Fluctuations in the data that occur over a longer, irregular period (several years), often influenced by economic cycles or external factors, not tied to a fixed calendar cycle.

📊 Example:

The business cycle of a country where periods of economic expansion are followed by recessions, each lasting several years.

4️⃣ Irregular (Random/Noise) Component (I)

🔍 Definition:
Random, unpredictable variations caused by one-off events or noise in data. These don’t follow a pattern.

📊 Example:

A sudden drop in stock prices due to an unexpected natural disaster or political event.

5. Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in
terms of structure and application.
- 1️⃣ AR (Auto-Regressive) Model

 🔍 Structure:

  Predicts the current value of the time series based on its past values (lags).
  Use Cases:

  Suitable for stationary time series where past values have a strong relationship with the current value.

  Example: Predicting stock prices based on past prices when trend and seasonality are removed.

  2️⃣ MA (Moving Average) Model

🔍 Structure:

Models the current value as a linear combination of past forecast errors (residuals).
Use Cases:

Suitable when random shocks (noise) explain the time series dynamics.

Example: Forecasting economic indicators where irregular disturbances dominate

3️⃣ ARIMA (Auto-Regressive Integrated Moving Average)

🔍 Structure:

Combines AR and MA models with differencing (Integration) to handle non-stationary series.

Notation: ARIMA(p, d, q) where:

p: AR order.

d: Number of differences needed to make the series stationary.

q: MA order.

Use Cases:

Forecasting non-stationary time series after differencing.

Example: Predicting sales where trends exist, but no strong seasonality.

4️⃣ SARIMA (Seasonal ARIMA)

🔍 Structure:

Extends ARIMA by adding seasonal terms:

Seasonal AR (P), Seasonal MA (Q), and Seasonal differencing (D).

Notation: SARIMA(p, d, q)(P, D, Q, s):

s: Seasonal period (e.g., s=12 for monthly data with yearly seasonality).

📊 Use Cases:

Time series with seasonality + trend + noise.

Example: Monthly retail sales data showing yearly seasonal effects.

5️⃣ SARIMAX (SARIMA with eXogenous variables)

🔍 Structure:

SARIMA model that also includes exogenous variables (X) — additional predictors outside the time series itself.

Use Cases:

When external factors affect the time series.

Example: Forecasting electricity demand based on temperature, holidays, or promotions.

```python
# 6. Load a time series dataset (e.g., AirPassengers), plot the original series,and decompose it into trend, seasonality, and residual components.

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Step 1: Load the dataset
file_path = '/content/AirPassengers.csv'
data = pd.read_csv(file_path)

# Step 2: Prepare the time series
data['Month'] = pd.to_datetime(data['Month'])
data.set_index('Month', inplace=True)
time_series = data['#Passengers']

# Step 3: Plot the original time series
plt.figure(figsize=(12, 6))
plt.plot(time_series, label='Original Time Series')
plt.title('Air Passengers Time Series')
plt.xlabel('Date')
plt.ylabel('Number of Passengers')
plt.legend()
plt.show()

# Step 4: Decompose the time series
decomposition = seasonal_decompose(time_series, model='multiplicative', period=12)

# Step 5: Plot decomposition components
plt.figure(figsize=(14, 10))

plt.subplot(411)
plt.plot(time_series, label='Original', color='blue')
plt.legend(loc='upper left')

plt.subplot(412)
plt.plot(decomposition.trend, label='Trend', color='orange')
plt.legend(loc='upper left')

plt.subplot(413)
plt.plot(decomposition.seasonal, label='Seasonality', color='green')
plt.legend(loc='upper left')

plt.subplot(414)
plt.plot(decomposition.resid, label='Residual', color='red')
plt.legend(loc='upper left')

plt.tight_layout()
plt.show()
```

```python
# 8. Train a SARIMA model on the monthly airline passengers dataset.Forecast the next 12 months and visualize the results.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Load dataset
file_path = '/content/AirPassengers.csv'
data = pd.read_csv(file_path)
data['Month'] = pd.to_datetime(data['Month'])
data.set_index('Month', inplace=True)
time_series = data['#Passengers']

# Train SARIMA model
model = SARIMAX(time_series,
                order=(1, 1, 1),
                seasonal_order=(1, 1, 1, 12),
                enforce_stationarity=False,
                enforce_invertibility=False)

results = model.fit()

# Forecast next 12 months
forecast_steps = 12
forecast = results.get_forecast(steps=forecast_steps)
forecast_index = pd.date_range(start=time_series.index[-1] + pd.DateOffset(months=1), periods=forecast_steps, freq='MS')
forecast_series = pd.Series(forecast.predicted_mean.values, index=forecast_index)

# Get confidence intervals and set correct index
forecast_ci = forecast.conf_int()
forecast_ci.index = forecast_index  # Fix index to match forecast series

# Plot historical data and forecast
plt.figure(figsize=(14, 7))
plt.plot(time_series, label='Historical Data', color='blue')
plt.plot(forecast_series, label='Forecast (Next 12 Months)', color='orange')
plt.fill_between(forecast_ci.index,
                 forecast_ci['lower #Passengers'],
                 forecast_ci['upper #Passengers'],
                 color='pink', alpha=0.3, label='Confidence Interval')
plt.title('SARIMA Forecast of Air Passengers')
plt.xlabel('Date')
plt.ylabel('Number of Passengers')
plt.legend()
plt.show()
```

```python
# 9. Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib.
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

# Step 1: Load the dataset
file_path = '/content/AirPassengers.csv'
data = pd.read_csv(file_path)

# Step 2: Prepare the time series
data['Month'] = pd.to_datetime(data['Month'])
data.set_index('Month', inplace=True)
time_series = data['#Passengers']

# Step 3: Prepare data for LOF (needs 2D array)
X = time_series.values.reshape(-1, 1)

# Step 4: Apply Local Outlier Factor (LOF)
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
y_pred = lof.fit_predict(X)  # -1 for outliers, 1 for inliers

# Step 5: Extract anomalies
anomalies = time_series[y_pred == -1]

# Step 6: Visualize the results
plt.figure(figsize=(14, 7))
plt.plot(time_series, label='Air Passengers', color='blue')
plt.scatter(anomalies.index, anomalies.values, color='red', label='Anomalies', marker='o')
plt.title('Anomaly Detection in Air Passengers using LOF')
plt.xlabel('Date')
plt.ylabel('Number of Passengers')
plt.legend()
plt.show()
```
